```python
import time
from google.genai.types import Content, Part

def upload_vidoe(video_file_name):
  video_file = client.files.upload(path=video_file_name)
  while video_file.state == "PROCESSING":
    print("Waiting for video to be processed")
    time.sleep(10)
    video_file = client.files.get(name=video_file.name or "")

  if video_file.state == "FAILED":
    raise ValueError(video_file.state)
  print (f"video processing complete : " + video_file.uri or "")
  return video_file

# Call the upload_vidoe function and assign the result to my_video
video_file_name = "/content/WhatsApp Video 2024-12-28 at 10.22.00 AM.mp4"
my_video = upload_vidoe(video_file_name)

promt = """For each scene in this video, Generate captions that describe the scene along
            with any spoken text placed in quotation marks.
            Place each captions into an object wiht hte timecode of the caption in the vidoe.
            And write clear name of the same at the end."""

# Now you can use my_video
video = my_video

response = client.models.generate_content(
    model=model,
    contents=[
        Content(
            role="user",
            parts=[
                Part.from_uri( #Changed part to Part
                    file_uri=video.uri or "",
                    mime_type=video.mime_type or ""),
              ]),
        promt,
            ]
        )
display(response.text)
```
